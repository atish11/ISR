```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

filename = "C:/Users/HP/OneDrive/Desktop/ISR/Assign 5/download.png"

img = cv2.imread(filename)                   # <-- add this
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   # <-- add this

plt.figure(figsize=(5,5))
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
colors = ('r', 'g', 'b')
plt.figure(figsize=(8,5))
for i, color in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)
    plt.xlim([0, 256])

plt.title("Color Histogram (Feature Representation)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
mean_colors = cv2.mean(img)[:3]
print(f"Mean R: {mean_colors[0]:.2f}, G: {mean_colors[1]:.2f}, B: {mean_colors[2]:.2f}")
```

```python

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import string
```

```python
nltk.download('punkt')  
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
filename = "C:/Users/HP/OneDrive/Desktop/ISR/Assign 1 conflation/Conflation.txt"  # <-- use your own text file
with open(filename, 'r', encoding='utf-8') as file:
    text = file.read()
```

```python
print("Original Text:\n", text)
print("-" * 80)
```

```python
tokens = word_tokenize(text.lower())
```

```python
stop_words = set(stopwords.words('english'))
tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
```

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens]
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]
```

```python
print("After Stemming:\n", ' '.join(stemmed_words))
print("-" * 80)
print("After Lemmatization:\n", ' '.join(lemmatized_words))
print("-" * 80)
```

```python
from collections import Counter
freq = Counter(stemmed_words)
print("Document Representative (Word Frequency):")
for word, count in freq.most_common(10):
    print(f"{word}: {count}")
```
